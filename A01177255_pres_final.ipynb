# Pregunta de Interes 2
* ¿Cual seria el impacto de 'Age' , 'Wins' , 'TFights' , 'Losses' , 'OLosses' en 'Purse'? 
* Purse = b0 + b1*Age + b2*Wins + b3*TFights + b4*OWins + b5*OLosses

```python
df.corr()
```

```python
modelA = sm.OLS.from_formula('Purse~Age+Wins+TFights+OWins +OLosses', data = df)
resultA = modelA.fit()
resultA.summary() 
```

### Observación: 
* R-squared: 0.062; por lo tanto 6.2% de Purse es explicado por el modelo
* En el modelo vemos la correlación con Purse que Age: 0.121, Wins: 0.211, TFights:0.169, OLosses: -0.0061, Owins:0.197
Por lo que todas estas variables afectan positivamente a Purse excepto por OLosses
